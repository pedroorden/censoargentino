## 3. `comparar()` — pivot geográfico

```python
# NBI por provincia — el caso más común
censo.comparar("HOGAR_NBI_TOT")
```

```python
# Sexo por provincia
censo.comparar("PERSONA_P02")
```

```python
# IPMH por provincia
censo.comparar("HOGAR_IPMH")
```

```python
# NBI por departamento dentro de una provincia
censo.comparar("HOGAR_NBI_TOT", nivel="departamento", provincia="Tucumán")
```

```python
# Educación por departamento — Córdoba
censo.comparar("PERSONA_MNI", nivel="departamento", provincia="Córdoba")
```

---
## 4. `agregar()` — función pura sobre un DataFrame existente

```python
# Primero traemos datos crudos
df = censo.query(variables="PERSONA_P02", provincia="Santa Fe")
print(f"Filas crudas: {len(df):,}")
df.head(3)
```

```python
# Totales provinciales (Santa Fe)
censo.agregar(df)
```

```python
# Abierto por departamento
censo.agregar(df, por="departamento")
```

```python
# Mismo df, distinta agregación — sin volver a descargar
por_prov = censo.agregar(df, por="provincia")
por_prov
```

---
## 5. Composición — `query()` + `agregar()` manual

```python
# Hacinamiento en Mendoza, abierto por departamento
df_hac = censo.query(variables="HOGAR_H20CP", provincia="Mendoza")
censo.agregar(df_hac, por="departamento")
```

---
## 6. Manejo de errores

```python
# Provincia inválida
try:
    censo.tabla("PERSONA_P02", provincia="Atlantida")
except ValueError as e:
    print(f"ValueError: {e}")
```

```python
# Provincia ambigua
try:
    censo.tabla("PERSONA_P02", provincia="san")
except ValueError as e:
    print(f"ValueError: {e}")
```

```python
# nivel inválido en comparar()
try:
    censo.comparar("HOGAR_NBI_TOT", nivel="municipio")
except ValueError as e:
    print(f"ValueError: {e}")
```

```python
import pandas as pd

# Verificar que tabla() devuelve siempre un DataFrame con la estructura correcta
resultado = censo.tabla("PERSONA_P02", provincia="Córdoba")
assert isinstance(resultado, pd.DataFrame)
assert list(resultado.columns) == ["categoria", "N", "%"]
print("Estructura OK:", list(resultado.columns))
print(f"Filas: {len(resultado)}")
resultado
```
